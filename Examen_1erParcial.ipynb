```python
#Ejercicio 1

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Datos
X = np.array([1,3]).reshape(-1,1) #variables independientes
Y = np.array([2,2]) #variables dependientes

#modelo
modelo = LinearRegression()
modelo.fit(X,Y)

#Parametros
m = modelo.coef_[0]
b = modelo.intercept_

print("Pendiente: ", m)
print("Intercepto: ", b)
print(f"ecuación: y= {m}x + {b}")

#Graficar
x_vals = np.linspace(0,4,100).reshape(-1,1)
y_vals = modelo.predict(x_vals)

plt.scatter(X,Y)
plt.plot(x_vals, y_vals)
plt.title("recta que pasa por (1,2) y (3,2)")
plt.show()
```
