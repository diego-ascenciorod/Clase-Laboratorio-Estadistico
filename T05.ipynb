```python
# regresi贸n mpg

import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Cargar los datos
df = pd.read_excel("Motor Trend Car Road Tests (2).xlsx")

# eliminar columna model
if "model" in df.columns:
    df = df.drop("model", axis=1)

# definir X y Y
X = df.drop("mpg", axis=1)
y = df["mpg"]

print("shape:", df.shape)
print("columnas:", df.columns.tolist())
```

```python
# modelo lineal
lr = LinearRegression()
lr.fit(X, y)

# R2 total
R2_total = lr.score(X, y)
print("R2 total:", R2_total)

# betas
betas = pd.DataFrame({
    "variable": X.columns,
    "beta": lr.coef_
})
print("\nbetas:")
print(betas)
```

```python
# train-test split 40% entrenamiento
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.4, random_state=42
)

lr_split = LinearRegression()
lr_split.fit(X_train, y_train)

print("R2 entrenamiento:", lr_split.score(X_train, y_train))
print("R2 prueba:", lr_split.score(X_test, y_test))
```

```python
# estandarizaci贸n
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_scaled, y, train_size=0.4, random_state=42
)

print("ridge mpg")
for alpha in [0.1, 1, 10, 100]:

    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train_s, y_train_s)

    print("alpha:", alpha)
    print("R2 train:", ridge.score(X_train_s, y_train_s))
    print("R2 test:", ridge.score(X_test_s, y_test_s))
```

```python
# regresi贸n qsec
# definir X e y
X_q = df.drop("qsec", axis=1)
y_q = df["qsec"]

print("shape X:", X_q.shape)
```

```python
# modelo lineal
lr_q = LinearRegression()
lr_q.fit(X_q, y_q)

# R2 total
R2_total_q = lr_q.score(X_q, y_q)
print("R2 total:", R2_total_q)

# betas
betas_q = pd.DataFrame({
    "variable": X_q.columns,
    "beta": lr_q.coef_
})

print("\nbetas:")
print(betas_q)
```

```python
# train-test split
X_train_q, X_test_q, y_train_q, y_test_q = train_test_split(
    X_q, y_q, train_size=0.4, random_state=42
)

lr_split_q = LinearRegression()
lr_split_q.fit(X_train_q, y_train_q)

print("R2 entrenamiento:", lr_split_q.score(X_train_q, y_train_q))
print("R2 prueba:", lr_split_q.score(X_test_q, y_test_q))
```

```python
# estandarizaci贸n
scaler_q = StandardScaler()
X_q_scaled = scaler_q.fit_transform(X_q)

X_train_s_q, X_test_s_q, y_train_s_q, y_test_s_q = train_test_split(
    X_q_scaled, y_q, train_size=0.4, random_state=42
)

print("ridge qsec")

for alpha in [0.1, 1, 10, 100]:

    ridge_q = Ridge(alpha=alpha)
    ridge_q.fit(X_train_s_q, y_train_s_q)

    print("alpha:", alpha)
    print("R2 train:", ridge_q.score(X_train_s_q, y_train_s_q))
    print("R2 test:", ridge_q.score(X_test_s_q, y_test_s_q))
```

```python
# mpg con dummies
# crear dummies para cyl, gear y carb
df_d = pd.get_dummies(df, columns=["cyl", "gear", "carb"], drop_first=True)

# definir X e y
X_d = df_d.drop("mpg", axis=1)
y_d = df_d["mpg"]

print("shape:", df_d.shape)
print("columnas nuevas:")
print(df_d.columns.tolist())
```

```python
lr_d = LinearRegression()
lr_d.fit(X_d, y_d)

print("R2 total:", lr_d.score(X_d, y_d))

betas_d = pd.DataFrame({
    "variable": X_d.columns,
    "beta": lr_d.coef_
})

print("\nbetas:")
print(betas_d)
```

```python
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_d, y_d, train_size=0.4, random_state=42
)

lr_split_d = LinearRegression()
lr_split_d.fit(X_train_d, y_train_d)

print("R2 entrenamiento:", lr_split_d.score(X_train_d, y_train_d))
print("R2 prueba:", lr_split_d.score(X_test_d, y_test_d))
```

```python
# qsec con dummies

X_dq = df_d.drop("qsec", axis=1)
y_dq = df_d["qsec"]

lr_dq = LinearRegression()
lr_dq.fit(X_dq, y_dq)

print("R2 total:", lr_dq.score(X_dq, y_dq))

betas_dq = pd.DataFrame({
    "variable": X_dq.columns,
    "beta": lr_dq.coef_
})

print("\nbetas:")
print(betas_dq)

X_train_dq, X_test_dq, y_train_dq, y_test_dq = train_test_split(
    X_dq, y_dq, train_size=0.4, random_state=42
)

lr_split_dq = LinearRegression()
lr_split_dq.fit(X_train_dq, y_train_dq)

print("R2 entrenamiento:", lr_split_dq.score(X_train_dq, y_train_dq))
print("R2 prueba:", lr_split_dq.score(X_test_dq, y_test_dq))
```
