```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import files

datos = pd.read_csv("Salarios_minimos.csv")

X = datos["Periodo"].values      # variable explicativa
y = datos["Salario"].values      # variable respuesta

X_mat = X.reshape(-1, 1)

lr = LinearRegression()
lr.fit(X_mat, y)

beta_0 = lr.intercept_
beta_1 = lr.coef_[0]

print("β0 =", beta_0)
print("β1 =", beta_1)

y_hat = lr.predict(X_mat)

R2 = lr.score(X_mat,y)
print("R2 =", R2)

n = len(y)
p = 2

RSS = np.sum((y - y_hat)**2)
var = RSS / (n - p)
print("Varianza =", var)

x_bar = np.mean(X)

aux = np.sum((X - x_bar)**2)

SE_beta1 = np.sqrt(var / aux)
print("SE(β1) =", SE_beta1)

SE_beta0 = np.sqrt(
    var * (1/n + (x_bar**2)/aux)
)

print("SE(β0) =", SE_beta0)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Se va a analizar el Boston Housing Dataset con el objetivo de"
"estudiar la relación entre distintas variables socioeconómicas y el valor medio"
"de las viviendas (MEDV).")

df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
df.head()
df.shape
df.info()
df.describe()

df.corr()["MEDV"].sort_values(ascending=False)

X = df.drop(columns=["MEDV"])
y = df["MEDV"]


X_mm = (X - X_min) / (X_max - X_min)
X_mm.describe()

X_mean = X.mean()
X_std = X.std()

X_standard = (X - X.mean()) / X.std()


X_standard.describe()
plt.hist(X_standard)
plt.xlabel("RM estandarizado")
plt.ylabel("Frecuencia")
plt.title("Histograma de RM (Standard Scaling)")
plt.show()
```
