```python
# ejercicio 3

# probabilidad inicial de lluvia (pronóstico)
P_lluvia = 0.25
P_no_lluvia = 1 - P_lluvia

# cada amigo miente con probabilidad 1/3
P_mentir = 1/3
P_decir_verdad = 1 - P_mentir

# si está lloviendo, los 3 dicen la verdad
P_dicen_lluvia_dado_lluvia = P_decir_verdad**3

# si no está lloviendo, los 3 tendrían que mentir
P_dicen_lluvia_dado_no_lluvia = P_mentir**3

# probabilidad total de que los tres digan que llueve
P_dicen_lluvia = (
    P_dicen_lluvia_dado_lluvia * P_lluvia +
    P_dicen_lluvia_dado_no_lluvia * P_no_lluvia
)

# bayes
P_lluvia_dado_dicen = (
    P_dicen_lluvia_dado_lluvia * P_lluvia
) / P_dicen_lluvia

print("Probabilidad de que sí esté lloviendo:", P_lluvia_dado_dicen)
```

```python
# ejercicio 4

import numpy as np

np.random.seed(0)

# poblaciones
mujeres = 185392
hombres = 162690
otros = 30268

total = mujeres + hombres + otros

# probabilidades de cada grupo
p_mujeres = mujeres / total
p_hombres = hombres / total
p_otros = otros / total

# parámetros de peso
mu_mujeres = 70
sd_mujeres = 11

mu_hombres = 80
sd_hombres = 16

mu_otros = 73
sd_otros = 12

# simulación
n_sim = 10000
excesos = 0

for _ in range(n_sim):

    peso_total = 0

    for _ in range(4):

        grupo = np.random.choice(
            ["mujer","hombre","otro"],
            p=[p_mujeres, p_hombres, p_otros]
        )

        if grupo == "mujer":
            peso = np.random.normal(mu_mujeres, sd_mujeres)
        elif grupo == "hombre":
            peso = np.random.normal(mu_hombres, sd_hombres)
        else:
            peso = np.random.normal(mu_otros, sd_otros)

        peso_total += peso

    if peso_total > 300:
        excesos += 1

print("Veces que se excede el peso:", excesos)
```
