```python
# librerias
import numpy as np
import matplotlib.pyplot as plt

# datos iniciales
x = np.array([10, 2, 3])
y = np.array([2847, -4017, -1255])

# intervalo de busqueda
xmin = 2
xmax = 15

# mostrar los datos
print("valores iniciales")
for i in range(len(x)):
    print("x =", x[i], " y =", y[i])

# mejor valor actual
pos_mejor = np.argmin(y)
x_mejor = x[pos_mejor]
y_mejor = y[pos_mejor]

print("\nmejor valor actual")
print("x min =", x_mejor)
print("y min =", y_mejor)

# grafica de los puntos
plt.figure(figsize=(8,5))
plt.scatter(x, y, s=80)
plt.title("valores iniciales")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

# crear puntos candidatos dentro del intervalo
x_candidatos = np.arange(xmin, xmax + 1, 1)

print("candidatos en el intervalo")
print(x_candidatos)

# quitar los que ya se probaron
x_nuevos = []
for valor in x_candidatos:
    if valor not in x:
        x_nuevos.append(valor)

x_nuevos = np.array(x_nuevos)

print("\nvalores que aun no se han probado")
print(x_nuevos)

# elegir el mas cercano al mejor punto actual
distancias = abs(x_nuevos - x_mejor)
pos_siguiente = np.argmin(distancias)
x_siguiente = x_nuevos[pos_siguiente]

print("\nsiguiente valor sugerido")
print("x_next =", x_siguiente)

# grafica marcando el siguiente punto sugerido
plt.figure(figsize=(8,5))
plt.scatter(x, y, s=80, label="valores iniciales")
plt.axvline(x=x_siguiente, linestyle="--", label="siguiente x sugerida")
plt.title("busqueda del siguiente punto")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# librerias
import numpy as np
import matplotlib.pyplot as plt

# semilla aleatoria (para que siempre salga igual)
np.random.seed(42)

# funcion real
def f(x):
    return np.sin(x) + 0.5*x

# intervalo de busqueda
x_min = 2
x_max = 15

# numero de puntos
n = 10001

# generar puntos en el intervalo
x_grid = np.linspace(x_min, x_max, n)

# evaluar funcion
y = f(x_grid)

# agregar ruido
epsilon = np.random.normal(0, 0.5, size=n)
y_hat = y + epsilon

# encontrar minimo
pos = np.argmin(y_hat)
x_best = x_grid[pos]
y_best = y_hat[pos]

print("mejor x encontrada:", x_best)
print("valor minimo:", y_best)

# grafica
plt.figure(figsize=(8,5))

plt.plot(x_grid, y, label="funcion real")
plt.scatter(x_grid, y_hat, s=2, alpha=0.3, label="datos con ruido")
plt.scatter(x_best, y_best, color="red", label="minimo encontrado")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# instalar libreria
!pip -q install scikit-learn

# librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# semilla
random_state = 42
np.random.seed(random_state)

# datos iniciales
X = np.array([10, 2, 3, 7, 12]).reshape(-1, 1)
y = np.array([2847, -4017, -1255, 1773, 563], dtype=float)

# mostrar tabla inicial
print("tabla inicial")
for i in range(len(X)):
    print("x =", X[i, 0], " y =", y[i])

# intervalo de busqueda
x_min = 2
x_max = 15
n = 10001

# puntos de prueba
X_test = np.linspace(x_min, x_max, n).reshape(-1, 1)

# modelo gpr
kernel = RBF(length_scale=1.0)
gp = GaussianProcessRegressor(kernel=kernel, random_state=random_state)

# entrenar modelo
gp.fit(X, y)

# predecir
y_hat, sigma = gp.predict(X_test, return_std=True)

# limites
y_upper = y_hat + 2 * sigma
y_lower = y_hat - 2 * sigma

# encontrar el x donde y_upper es maximo
idx = np.argmax(y_upper)
x_new = X_test[idx, 0]
y_new = y_hat[idx]

print("\nnuevo punto sugerido")
print("x_new =", x_new)
print("y_new =", y_new)

# agregar nuevo punto
X = np.vstack([X, [[x_new]]])
y = np.append(y, y_new)

print("\ntabla actualizada")
for i in range(len(X)):
    print("x =", X[i, 0], " y =", y[i])

# grafica
plt.figure(figsize=(10, 6))
plt.scatter(X[:-1], y[:-1], s=80, label="datos iniciales")
plt.plot(X_test, y_hat, label="y_hat")
plt.plot(X_test, y_upper, "--", label="y_upper = y_hat + 2*sigma")
plt.plot(X_test, y_lower, "--", label="y_lower = y_hat - 2*sigma")
plt.scatter(x_new, y_new, s=100, label="nuevo punto")
plt.xlabel("x")
plt.ylabel("y")
plt.title("bayesian optimization con gpr")
plt.grid(True)
plt.legend()
plt.show()
```
