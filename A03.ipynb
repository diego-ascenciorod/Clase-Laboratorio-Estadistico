```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("aguasdelfeo.csv")
df["tamano"] = df["tamano"].str.lower().str.strip()
df["hielo"] = df["hielo"].str.lower().str.strip()
df["sabor"] = df["sabor"].str.lower().str.strip()

# chica = 1, grande = -1
df["tamano"] = df["tamano"].map({"chica": 1, "chico": 1, "grande": -1})

df["hielo"] = df["hielo"].map({"sin": -1, "con": 1})

dummies = pd.get_dummies(df["sabor"]).astype(int)
df = pd.concat([df.drop("sabor", axis=1), dummies], axis=1)

df = df[["tamano", "hielo", "kahlua", "horchata", "score"]]
print("Tabla:")
print(df)

X = df[["tamano", "hielo", "kahlua", "horchata"]]
y = df["score"]

modelo = sm.OLS(y, X).fit()

print("\nBetas con codificación -1 / 1:")
print(modelo.params)
```
